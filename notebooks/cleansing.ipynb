```python
import pandas as pd

# 1. Load the raw dataset
df = pd.read_csv('../data/DATASET.csv')

# 2. Handle Outliers (Capping Method)
# As we discussed, we don't want to delete medical data, but we can "cap" 
# extreme values so they don't confuse the Random Forest.
for feature in ['creatinine_phosphokinase', 'serum_creatinine']:
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    
    # Any value higher than the upper bound gets set to the upper bound
    df[feature] = df[feature].clip(upper=upper_bound)

# 3. Save the clean version
# This ensures your raw data stays safe and your teammates can use the clean version
df.to_csv('../data/cleaned_dataset.csv', index=False)

print("Data cleansing complete! 'cleaned_dataset.csv' has been created in your data folder.")
```
